```python
X, Y = np.meshgrid(np.linspace(-1,1, 2001),np.linspace(-1,1, 2001))

# Test Franke values are between 0 and np.
# https://www.sfu.ca/~ssurjano/franke2d.html

X = (X + 1.0)*0.5
Y = (Y + 1.0)*0.5
result =  0.75*np.exp(-((9*X - 2)**2 )/4 -((9*Y - 2)**2 )/4)
result += 0.75*np.exp(-((9*X + 1)**2 )/49. -((9*Y + 1))/10)
result +=  0.5*np.exp(-((9*X - 7)**2 )/4 -((9*Y - 3)**2 )/4)
result +=  -0.2*np.exp(-((9*X - 4)**2 ) -((9*Y - 7)**2 ))
result = result * np.pi / np.max(result) - 0.002

print np.all(result<np.pi), np.all(result>0.)
print np.min(result), np.max(result)

plt.figure()
im = plt.imshow(result)
plt.colorbar(im)
plt.show()
```
